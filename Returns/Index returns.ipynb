```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### index vs index

```python
data_index = pd.read_csv("Indices_Data_01.csv", index_col = "Date")
data_index = data_index.drop(['1996-12-31']) #1 index data missing in first row
```

```python
data_index.head()
```

```python
#data.tail()
```

#### base 100

```python
base100 = (data_index / data_index.iloc[0] * 100)
#base100.head()
```

```python
#base100.tail()
```

```python
base100.plot(figsize=(15,6))
plt.show()
```

#### index returns

```python
ind_returns = (data_index / data_index.shift(1)) -1 
ind_returns.tail()
```

```python
annual_ind_returns = ind_returns.mean() * 250
print(annual_ind_returns)
```
