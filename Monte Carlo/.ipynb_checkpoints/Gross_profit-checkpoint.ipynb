```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_m = 170
rev_stdev = 20
iterations = 1000
```

#### generate 1000 revenue forecasts using normal distribution mean 170 stdev 20

```python
rev = np.random.normal(rev_m, rev_stdev, iterations)
rev
```

```python
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show()
```

#### generate 1000 COGS absolute value forecasts. COGS as % of sales is normal distribution with mean 60% and stdev 10%

```python
COGS = - (rev * np.random.normal(0.6, 0.1))
plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```

#### generate gross profit forecasts

```python
gross_profit = rev + COGS
gross_profit
```

```python
plt.figure(figsize=(15,6))
plt.plot(gross_profit)
plt.show()
```

#### build histogram

```python
plt.figure(figsize=(10,6))
plt.hist(gross_profit, bins = 30)
plt.show()
```
