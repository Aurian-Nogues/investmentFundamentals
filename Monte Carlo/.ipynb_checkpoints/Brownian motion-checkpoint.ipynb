#### r = log return of share price between yesterday and today
#### price today = price yesterday x e(r)

#### r is unknown, we calculate using brownian motion algo:
#### components are:
####        drift = direction of returns in the past = average return - 0.5 x variance = expected daily return
####        volatility = standard deviation x Z(rand(0;1)) with Z number of standard dev away from the mean

#### price today = price yesterday x e(drift + volatility)

#### repeat 1000 times



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
```

```python
data= pd.read_csv("PG_2007_2017.csv", index_col = "Date")
```

```python
data.head()
```

```python
log_returns = np.log(1 + data.pct_change())
log_returns.tail()
```

```python
data.plot(figsize=(10,6))
```

```python
log_returns.plot(figsize = (10,6))
```

```python
u = log_returns.mean()
u
```

```python
var = log_returns.var()
var
```

```python
drift = u - (0.5 * var)
drift
```

```python
stdev = log_returns.std()
stdev
```

```python
type(drift)
```

```python
norm.ppf(0.95) #if an event has 95% chance to happen it is within 1.65 stdev from the mean
```

```python
x = np.random.rand(10,2)
x
```

```python
norm.ppf(x)
```

```python
Z  = norm.ppf(np.random.rand(10,2))
```

```python
t_intervals = 1000
iterations = 100
```

```python
daily_returns = np.exp(drift.values + stdev.values * norm.ppf(np.random.rand(t_intervals, iterations)))
```

```python
daily_returns
```

```python
S0 = data.iloc[-1]
S0
```

```python
price_list = np.zeros_like(daily_returns)
```

```python
price_list
```

```python
price_list[0] = S0
```

```python
for t in range(1, t_intervals):
    price_list[t] = price_list[t-1] * daily_returns[t]
```

```python
price_list
```

```python
plt.figure(figsize=(10,6))
plt.plot(price_list)
plt.show()
```
