## Black Scholes

$$
d_1 = \frac{\ln(\frac{S}{K}) + (r + \frac{stdev^2}{2})t}{s \cdot \sqrt{t}}
$$

$$
d_2 = d_1 - s \cdot \sqrt{t} = \frac{\ln(\frac{S}{K}) + (r - \frac{stdev^2}{2})t}{s \cdot \sqrt{t}}
$$

```python
def d1(S, K, r, stdev, T):
    return (np.log(S / K) + (r + stdev **2 / 2) * T) / (stdev * np.sqrt(T))
def d2(S, K, r, stdev, T):
    return (np.log(S / K) + (r - stdev ** 2/2) * T) / (stdev * np.sqrt(T))
```

```python
norm.cdf(0) # cumulative normal distribution -> 50% of datas is below 0 which is the mean of the normal distribution
```

$$
\textbf{C} = SN(d_1) - Ke^{-rt}N(d_2) 
$$

```python
def BSM(S, K, r, stdev, T):
        return (S * norm.cdf(d1(S, K, r, stdev, T))) - (K * np.exp(-r * T) * norm.cdf(d2(S, K, r, stdev, T)))
```

### Rational: compute call option price using monte carlo simulation to generate future payoffs and discount to today

```python
data= pd.read_csv("PG_2007_2017.csv", index_col = "Date")
```

```python
log_returns = np.log(1 + data.pct_change())
```

```python
r = 0.025 # risk free rate
```

```python
stdev = log_returns.std() * 250 ** 0.5
stdev
```

```python
stdev = stdev.values #converts from dataframe to numpy array
```

```python
T = 1 #time horizon
t_intervals = 250
delta_t = T / t_intervals

iterations = 10000
```

```python
Z = np.random.standard_normal((t_intervals + 1, iterations))    # random component
S = np.zeros_like(Z)
S0 = data.iloc[-1] # current price
S[0] = S0

S0
```

```python
for t in range(1, t_intervals + 1):
    S[t] = S[t-1] * np.exp((r - 0.5 * stdev ** 2) * delta_t + stdev * delta_t ** 0.5 * Z[t])
```

```python
S
```

```python
plt.figure(figsize=(10,6))
plt.plot(S[:, :10]) # only show first 10 simulations
plt.show()
```
